```python
from pandas import read_csv
file='diabetes.csv'
columns = ['preg','Glu','b_pres','skin','insu', 'bmi','pedi','age','class']
df=read_csv(file, names=columns)
print(df.shape)
```

```python
correlation = df.corr()  # Calculate the correlation matrix for the dataframe

import numpy as np
fig = plt.figure()  # Create a new figure
ax = fig.add_subplot(111)  # Add a subplot to the figure
cax = ax.matshow(correlation, vmin=-1, vmax=1)  # Display the correlation matrix as a color-coded matrix
fig.colorbar(cax)  # Add a colorbar to the figure
ticks = np.arange(0, 9, 1)  # Create an array of ticks for the x and y axes
ax.set_xticks(ticks)  # Set the x-axis ticks
ax.set_yticks(ticks)  # Set the y-axis ticks
ax.set_xticklabels(correlation.columns)  # Set the labels for the x-axis ticks
ax.set_yticklabels(correlation.columns)  # Set the labels for the y-axis ticks
plt.show()  # Display the plot
```

Normalization

```python
from sklearn.preprocessing import MinMaxScaler
array=df.values
x=array[:,0:8]
y=array[:,8]
scaler = MinMaxScaler(feature_range = (0,1))
rescaledx=scaler.fit_transform(x)

print("original Data:\n\n",df.values)
print("\nstandardized Data: \n\n",rescaledx[0:5,:])
sns.boxplot(data=rescaledx, orient ='v')
```

```python
from pandas import read_csv
from sklearn.preprocessing import StandardScaler

array=df.values

x=array[:,0:8]
y=array[:,8]

scalar=StandardScaler().fit(x)
standardx= scalar.transform(x)

print("Original Data: \n",df.values)
print("\n\nStandardized Data:\n ",standardx[0:5,:])
```

binarization

```python
from pandas import read_csv
from sklearn.preprocessing import Binarizer

array= df.values

x=array[:,0:8]
y=array[:,8]

binarizer=Binarizer(threshold=0.2).fit(x)
binaryx=binarizer.transform(x)

print("Original Data:\n\n", df.values)
print("\n\nNormalized Data:\n",binaryx[0:5,:])
```
